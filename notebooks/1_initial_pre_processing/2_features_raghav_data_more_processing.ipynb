```python
import dask.dataframe as dd

eaglei = dd.read_csv('../data/cleaned_data2.csv')
eaglei.head(20)
```

```python
len(eaglei)
```

```python
# Filter eaglei for years 2020, 2021, 2022
eaglei_filtered = eaglei[eaglei['Year'].isin([2020, 2021, 2022])]

# Compute the result (lazy evaluation)
eaglei_2020_2022 = eaglei_filtered.compute()

# Print the shape to verify the filtering
print(f"Shape of filtered eaglei dataframe: {eaglei_2020_2022.shape}")

# Display the first few rows
eaglei_2020_2022.head()
```

```python
#save eaglei_2020_2022 to parquet
eaglei_2020_2022.to_parquet('../data/eaglei_2020_2022.parquet')
```

```python
# Check unique years in eaglei_2020_2022
unique_years = eaglei_2020_2022['Year'].unique()
year_counts = eaglei_2020_2022['Year'].value_counts().sort_index()

print("Unique years in the filtered dataset:")
print(unique_years)
print("\nCount of records per year:")
print(year_counts)
```

```python
len(eaglei_2020_2022)
```

```python
eaglei_grouped = eaglei_2020_2022.groupby(['state', 'run_date', 'run_time'])['customers'].sum().reset_index()
```

```python
eaglei_grouped.head()
```

```python
import pandas as pd
```

```python
# Convert run_time to proper time format if needed
if isinstance(eaglei_grouped['run_time'].iloc[0], str):
    # Combine run_date and run_time into a single datetime column
    eaglei_grouped['datetime'] = pd.to_datetime(
        eaglei_grouped['run_date'].astype(str) + ' ' + 
        eaglei_grouped['run_time'].astype(str)
    )
else:
    # If run_time is already in a time format
    eaglei_grouped['datetime'] = pd.to_datetime(
        eaglei_grouped['run_date'].dt.strftime('%Y-%m-%d') + ' ' + 
        eaglei_grouped['run_time']
    )

# Display the result to verify
print("DataFrame with new datetime column:")
eaglei_grouped.head()
```

```python
# Drop the original run_date and run_time columns since we have datetime column
eaglei_grouped = eaglei_grouped.drop(columns=['run_time', 'run_date'])

eaglei_grouped.head(2)
```

```python
# Save the eaglei_grouped DataFrame to parquet
eaglei_grouped.to_parquet('../data/eaglei_grouped_2020_2022.parquet')
```
